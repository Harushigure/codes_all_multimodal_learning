```python
import cv2
import os

# Create or check if the target folder exists
if not os.path.exists("/Users/18666163123/Desktop/Lab/LiangLab/Data/Recording_chromatophore activity/Alice/frame_extract"):  # Check if the folder named "frame_extract" exists.
    os.makedirs("/Users/18666163123/Desktop/Lab/LiangLab/Data/Recording_chromatophore activity/Alice/frame_extract")          # Create the folder if it does not exist.

video_path = "/Users/18666163123/Desktop/Lab/LiangLab/Data/Recording_chromatophore activity/Alice/Test.MOV"  # Define the path to the input video (replace with actual video path).
cap = cv2.VideoCapture(video_path)  # Load the video file using cv2.VideoCapture.

frame_count = 0  # Initialize variable frame_count to track the number of saved frames.
current_frame = 0  # Initialize variable current_frame to track the current frame number being processed.

while frame_count < 400:  # Continue loop while the number of saved frames is less than the target total of 400 frames.
    ret, frame = cap.read()  # Read the next frame of the video. ret indicates whether the read was successful, frame contains the image data.
    if not ret:  # If ret is False, it indicates the video has ended or read failed.
        break     # Exit the loop.

    current_frame += 1  # Increment the current frame number counter.

    if current_frame % 5 == 0:  # Frame interval is 5 frames.
        save_path = f"/Users/18666163123/Desktop/Lab/LiangLab/Data/Recording_chromatophore activity/Alice/frame_extract/{frame_count:04d}.jpg"  # Define save path with naming format 0000.jpg, etc.
        cv2.imwrite(save_path, frame)  # Save the current frame as an image file.
        frame_count += 1  # Increment the saved frame counter.

cap.release()  # Release the video capture object and close the video file.
print("Completed. 400 frames have been saved to the folder.")  # Print completion message.
```
