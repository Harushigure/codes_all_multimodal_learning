```python
import os
from PIL import Image

# Define input folder path and output folder path
input_folder = '/Users/18666163123/Desktop/Lab/LiangLab/Data/Recording_chromatophore activity/Alice/20250521/frame_extract'
output_folder = '/Users/18666163123/Desktop/Lab/LiangLab/Data/Recording_chromatophore activity/Alice/20250521/frame_extract_png'

# Create output folder if it does not exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Traverse all files in the input folder
for filename in os.listdir(input_folder):
    # Check if the file is in JPG format
    if filename.lower().endswith('.jpg'):
        # Construct complete file path
        input_path = os.path.join(input_folder, filename)

        # Open image file
        image = Image.open(input_path)

        # Construct output filename (replace .jpg with .png)
        output_filename = os.path.splitext(filename)[0] + '.png'
        output_path = os.path.join(output_folder, output_filename)

        # Save converted image to output folder
        image.save(output_path)

print("Conversion completed")
```
